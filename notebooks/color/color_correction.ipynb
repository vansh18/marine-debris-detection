```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = r'../../data/original_data/images/vid_000002_frame0000016.jpg'
original_image = cv.imread(image_path, cv.IMREAD_COLOR)
original_image = cv.cvtColor(original_image, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(original_image)
plt.show()
```

```python
# Brightnen the image

def brighten_image(image, alpha=1.0, beta=0):
    new_image = cv.convertScaleAbs(image, alpha=alpha, beta=beta)
    return new_image
```

```python
# Sharpen the image a little bit

def sharpen_image(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    new_image = cv.filter2D(image, -1, kernel)
    return new_image
```

```python
brighten_image = brighten_image(original_image, alpha=1.5, beta=0)
plt.imshow(brighten_image)
plt.show()
```

```python
sharpen_image = sharpen_image(brighten_image)
plt.imshow(sharpen_image)
plt.show()
```
