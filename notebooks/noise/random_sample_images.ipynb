```python
import os
import random

# Path to the folder containing all images
dataset_path = r'../data/original_data/images'

# Get a list of all image file paths
all_images = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Ensure the dataset contains the expected number of images
if len(all_images) != 7212:
    print(f"Warning: Dataset contains {len(all_images)} images instead of 7212.")

# Randomly sample 100 to 200 images
sample_size = random.randint(100, 200)
sampled_images = random.sample(all_images, sample_size)

# Print or save the sampled file paths
print(f"Sampled {sample_size} images:")
for img_path in sampled_images:
    print(img_path)

# Optionally, save the sampled file paths to a text file
output_file = "sampled_images.txt"
with open(output_file, 'w') as f:
    for img_path in sampled_images:
        f.write(img_path + '\n')

print(f"Sampled image paths saved to {output_file}")
```
